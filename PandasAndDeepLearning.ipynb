```python
# Concat and merge operation (Similar to Union All (not Union) and Join of SQL)

matrix1 = {
    "Name": ['Ram', 'Syam', 'Hari'],
    "Age": [20, 19, 17],
    "Roll no": [1, 2, 3]
}
```

```python
import pandas as pd
```

```python
table1 = pd.DataFrame(matrix1)
table1
```

```python
matrix2 = {
    "Name": ['a', 'b', 'c'],
    "Age": [20, 21, 18],
    "Roll no": [1, 2, 3]
}
```

```python
table2 = pd.DataFrame(matrix2)
table2
```

```python
table = pd.concat([table1, table2], axis = 0)
# table = pd.concat([table2, table1], keys = ['second table', 'first table'])
table
```

```python
table_pivot = table.pivot_table(columns = "Roll no", values = ['Age'], aggfunc= "mean")
table_pivot
```

```python
matrix1 = {

    "Name": ['Ram', 'Syam', 'Hari', 'Gita'],
    "Age": [20, 19, 17, 21],
    "Roll no": [1, 2, 3, 5]
}

table1 = pd.DataFrame(matrix1)
```

```python
matrix2 = {

    "Name": ['Arik', 'Ram', 'Syam', 'Hari'],
    "Phone_Num": [1234, 2345, 3456, 4567]
}

table2 = pd.DataFrame(matrix2)
```

```python
merged_table1 = pd.merge(table1, table2, how = "left", on = "Name")
```

```python
merged_table2 = pd.merge(table2, table1, how = "right", on = "Name")
```

```python
merged_table1
```

```python
merged_table2
```

```python
merged_table3 = pd.merge(table1, table2, how = "inner", on = "Name")
merged_table3
```

```python
merged_table4 = pd.merge(table1, table2, how = "outer", on = "Name")
merged_table4
```

```python
matrix = {
    "naam": ['Ram', 'Syam', 'Hari', 'Sita'],
    "phn": [1220, 9029, 2392, 2329]
}

table2 = pd.DataFrame(matrix)
table2
merged_table = pd.merge(table1, table2, how = "right", left_on = 'Name', right_on= "naam")
merged_table
```

```python
matrix = {
    "naam": ['Ram', 'Syam', 'Hari', 'Sita'],
    "phn": [1220, 9029, 2392, 2329]
}

table2 = pd.DataFrame(matrix)
table2 = table2.rename(columns = {"naam": 'Name'})
merged_table = pd.merge(table1, table2, how = "right", on= "Name")
merged_table
```

```python
# import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = data.load_data()
```

```python
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
train_images.shape, train_labels.shape, test_images.shape, test_labels.shape
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28, 28)),
    keras.layers.Dense(256, activation = "relu"),
    keras.layers.Dense(10, activation = "softmax")
])
```

```python
model.summary()
```

```python
model.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])
```
