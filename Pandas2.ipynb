```python
df = {'Name': ['Ram','Ram','Shyam'], 'Age': [20,21,20], 'is_female': [False,False,False]}
```

```python
import pandas as pd
data = pd.DataFrame(df)

data[(data['Age'] == 20) & (data['Name'] == 'Ram')]
data[(data['Age'] == 20) | (data['Name'] == 'Shyam')]
```

```python
df = {'Name': ['Ram','Sita','Shyam'], 'Age': [20,21,20], 'is_female': [False,True,False]}
```

```python
import pandas as pd
data = pd.DataFrame(df)

data[data['Name'].isin(['Ram', 'Sita'])]
data[(data['Name'] == 'Ram') | (data['Name'] == 'Sita')]
```

```python
data[['Name','Age']]
```

```python
data.loc[data['Name'].isin(['Ram','Sita']), ['Name','Age']]
```

```python
data.loc[(data['Name'] == 'Ram') | (data['Name'] == 'Sita'), ['Name', 'Age']]
```

```python
data.loc[data['Name'] == 'Sita', ('Name', 'is_female')]
```

```python
data.iloc[0:2, 0:2]
```

```python
data.iloc[0::2,0::2]
```

```python
# Creating plots in pandas
data.plot.bar(color='red')
```
