```python
# Advanced Pandas
data = {
    'Name': ['Ram', 'Sita', 'Syam', 'Radha'],
    'Age': [25, 21, 25, 25],
    'Marks': [600, 100, 250, 400],
    'Gender': ['Male','Female','Male','Female']
}

df = pd.DataFrame(data)
```

```python
df['Age'].mean()
```

```python
df[['Age', 'Gender']].groupby('Gender').mean()
```

```python
df.groupby('Gender')['Age'].mean()
```

```python
df.sort_values('Age')
```

```python
df.sort_values(by = 'Age')
```

```python
df.sort_values(['Age', 'Gender'])
```

```python
df[df['Age'] == 21]
```

```python
df_subset = df.sort_index().groupby('Gender').head(1)
df_subset
```
