Artificial Intelligence (AI) - AI is the branch of computer science that deals with providing Intelligence to the machines.

Searching and Sorting algorithms are AI but not ML.

Searching - Eg: Linear Search, Binary Search, etc. (List Searching)

Graph Searching
1. Uniformed (Blind) Search - Eg: Depth-First Search (DFS), Breadth-First Search (BFS) and Uniform-Cost Search

2. Informed (Heuristics) Search - Eg: Best-First Search, Hill Climbing, Simulated Annealing, A* algorithm.

Sorting

Bubble sort, Merge sort, Heap sort, Quick sort, etc

Machine Learning (ML) - Machine Learning is the subset of AI, where the machine should learn from data and improve with experience. Data would be involved from here.

Libraries and Tools for ML:

1. Pandas - For Data Analysis (Provides the DataFrames), helps to read csv, excel files.

2. NumPy - Since, Python doesn't have general support for multidimensional arrays. Numerical Python help us on that.

3. Scikit-learn - Model Building library, also helps for preprocessing

4. Joblib, Pickle - helps to save the trained data/ model.

5. Matplotlib - Default visualization library of Python.

6. Seaborn - Helps to provide statistical charts

7. Bokeh - Helps to provide interactive charts/ HoverTools.

Types of Machine Learning:

1. Supervised Machine Learning (used on labeled datasets)

  a. Regression: Predicting a value with labeled data is called Regression. For eg: 20 Degree Centigrade temperature for tomorrow. Algorithms used are Linear Regression, Non-Linear Regression, etc.

  b. Classification: Predicting a class/ category with labeled dataset is called Classification. Eg: High or Low?, Foggy or Cloudy?, Algorithms used are Logistic Regression, K-Nearest Neighbors (KNN), Decision Tree, Naive Bayes, Support Vector Classification (svm.SVC), etc.

```python
# X = df.drop(columns = ['genre'])
X = df[['age', 'gender']]
Y = df['genre']
```

```python
# Decision Tree

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model
```

```python
model.fit(X.values,Y.values)
```

```python
model.predict([[21,1]])
```

```python
# KNN

import pandas as pd
df = pd.read_csv('/content/music.csv')

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
```

```python
model.fit(X.values, Y.values)
```

```python
model.predict([[26,0]])
```

```python
# Naive Bayes

import pandas as pd
df = pd.read_csv('/content/music.csv')

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X.values, Y.values)
model.predict([[26,0]])
```

```python
# SVM

import pandas as pd
df = pd.read_csv('/content/music.csv')

from sklearn.svm import LinearSVC
model = LinearSVC()
model.fit(X.values, Y.values)
model.predict([[26,0]])
```

```python
# Logistic Regression
import pandas as pd
df = pd.read_csv('/content/music.csv')

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter = 1000)
model.fit(X.values, Y.values)
model.predict([[26,0]])
```
